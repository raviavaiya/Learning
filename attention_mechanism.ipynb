```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample input sentence and translation
input_sentence = ["I", "am", "learning", "attention"]
output_translation = ["Je", "suis", "en", "train", "d'", "apprendre", "l'attention"]

# Simulate attention weights (random for example)
attention_weights = torch.rand(len(output_translation), len(input_sentence))  # shape: [target_len, source_len]

# Normalize (optional but helps visualization)
attention_weights = torch.softmax(attention_weights, dim=-1)

# Plot the attention weights as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(attention_weights.detach().numpy(), 
            xticklabels=input_sentence, 
            yticklabels=output_translation, 
            cmap='viridis', 
            cbar=True, annot=True)

plt.title("Attention Mechanism Visualization (English  French)")
plt.xlabel("Input (English)")
plt.ylabel("Output (French)")
plt.show()
```
